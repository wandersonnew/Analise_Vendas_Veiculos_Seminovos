**Previsão de Preço**

Prevendo preço pelo ano de fabricação e preço da venda

Está sendo feito uma análise geral

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
df = pd.read_csv("dataset/vendas_loja_seminovos.csv", sep=";")
```

```python
X = df[['Ano_fabricacao']]
y = df['Valor_venda']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
print(f"Previsões: {predictions}")
```

```python
print(f"Valores reais: {np.array(y_test)}")
mae = mean_absolute_error(y_test, predictions)
print(f"Erro Médio Absoluto (MAE): {mae}")
```

```python
novo_carro = pd.DataFrame({'Ano_fabricacao': [2024]})
predicted_price = model.predict(novo_carro)

print(f"Preço previsto para um carro ano 2025: R$ {predicted_price[0]:.2f}")
```
